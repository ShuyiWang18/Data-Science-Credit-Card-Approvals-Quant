+ Define `calc_iv` function to [calculate](https://www.kaggle.com/puremath86/iv-woe-starter-for-python) Information Value and WOE Value

### Binary Features

```python
# Calculate information value
def calc_iv(df, feature, target, pr=False):
    lst = []
    df[feature] = df[feature].fillna("NULL")

    for i in range(df[feature].nunique()):
        val = list(df[feature].unique())[i]
        lst.append([feature,                                                        # Variable
                    val,                                                            # Value
                    df[df[feature] == val].count()[feature],                        # All
                    df[(df[feature] == val) & (df[target] == 0)].count()[feature],  # Good (think: Fraud == 0)
                    df[(df[feature] == val) & (df[target] == 1)].count()[feature]]) # Bad (think: Fraud == 1)

    data = pd.DataFrame(lst, columns=['Variable', 'Value', 'All', 'Good', 'Bad'])
    data['Share'] = data['All'] / data['All'].sum()
    data['Bad Rate'] = data['Bad'] / data['All']
    data['Distribution Good'] = (data['All'] - data['Bad']) / (data['All'].sum() - data['Bad'].sum())
    data['Distribution Bad'] = data['Bad'] / data['Bad'].sum()
    data['WoE'] = np.log(data['Distribution Good'] / data['Distribution Bad'])

    data = data.replace({'WoE': {np.inf: 0, -np.inf: 0}})

    data['IV'] = data['WoE'] * (data['Distribution Good'] - data['Distribution Bad'])

    data = data.sort_values(by=['Variable', 'Value'], ascending=[True, True])
    data.index = range(len(data.index))

    if pr:
        print(data)
        print('IV = ', data['IV'].sum())

    iv = data['IV'].sum()
    print('This variable\'s IV is:',iv)
    print(df[feature].value_counts())
    return iv, data
```

```python
def convert_dummy(df, feature,rank=0):
    pos = pd.get_dummies(df[feature], prefix=feature)
    mode = df[feature].value_counts().index[rank]
    biggest = feature + '_' + str(mode)
    pos.drop([biggest],axis=1,inplace=True)
    df.drop([feature],axis=1,inplace=True)
    df=df.join(pos)
    return df
```

```python
def get_category(df, col, binsnum, labels, qcut = False):
    if qcut:
        localdf = pd.qcut(df[col], q = binsnum, labels = labels) # quantile cut
    else:
        localdf = pd.cut(df[col], bins = binsnum, labels = labels) # equal-length cut

    localdf = pd.DataFrame(localdf)
    name = 'gp' + '_' + col
    localdf[name] = localdf[col]
    df = df.join(localdf[name])
    df[name] = df[name].astype(object)
    return df
```

```python
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]

    print(cm)

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

## IVWOEConcept and Application

Weight of Evidence(WoE):  

$$wo{e_i} = \ln {{{P_{yi}}} \over {{P_{ni}}}} = \ln {{{y_i}/{y_s}} \over {{n_i}/{n_s}}}$$
$wo{e_i}$ is the I category's WOE value. ${{P_{yi}}}$ is the proportion of the positive samples in this category to all positive samples.   ${{P_{ni}}}$ is the ratio of negative samples (${{n_i}}$) in this class to all negative samples (${{n_s}}$).

Information Value (IV):  
$$I{V_i} = ({P_{yi}} - {P_{ni}}) \times wo{e_i}$$  
The IV values of the various types are the difference between the conditional positive rate and the conditional negative rate multiplied by the WOE value of the variable. The total IV value of the variable can be understood as the weighted sum of the conditional positive rate and the conditional negative rate difference:
$$IV = \sum\limits_i^n {I{V_i}} $$  

The IV value measures the variable's ability to predict.


Relationship between IV value and predictive power

| IV| Ability to predict |
|:------|:------:|
| <0.02 | Almost no predictive power |
|0.02~0.1 |weak predictive power|
|0.1~0.3|Moderate predictive power|
|0.3~0.5|Strong predictive power|
|>0.5|Predictive power is too strong, need to check variables|

## Random Wheel Classifier


```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix

# === RANDOM WHEEL CLASSIFIER ===
class RandomWheelClassifier:
    def __init__(self, n_estimators=10, max_depth=5, random_state=42):
        self.n_estimators = n_estimators
        self.max_depth = max_depth
        self.random_state = random_state
        self.models = []

    def fit(self, X, y):
        self.models = []
        np.random.seed(self.random_state)
        for i in range(self.n_estimators):
            idx = np.random.choice(len(X), len(X), replace=True)
            X_sample = X.iloc[idx]
            y_sample = y.iloc[idx]

            clf = DecisionTreeClassifier(max_depth=self.max_depth, random_state=self.random_state + i)
            clf.fit(X_sample, y_sample)
            self.models.append(clf)

    def predict(self, X):
        preds = np.array([model.predict(X) for model in self.models])
        final_preds = np.round(np.mean(preds, axis=0)).astype(int)
        return final_preds

    def predict_with_confidence(self, X):
        preds = np.array([model.predict(X) for model in self.models])
        confidence = np.mean(preds, axis=0)
        final_preds = (confidence >= 0.5).astype(int)
        return final_preds, confidence

    def explain_instance(self, X_row):
        explanations = []
        for model in self.models:
            importances = model.feature_importances_
            explanations.append(importances)
        avg_importance = np.mean(explanations, axis=0)
        return dict(zip(X_row.index, avg_importance))

# === CLEAN DATA ===
# Drop problematic columns like 'wkphone' if they exist
if 'wkphone' in X_train.columns:
    X_train_clean = X_train.drop(columns=['wkphone'])
    X_test_clean = X_test.drop(columns=['wkphone'])
else:
    X_train_clean = X_train.copy()
    X_test_clean = X_test.copy()

# === TRAIN THE MODEL ===
rw_model = RandomWheelClassifier(n_estimators=15, max_depth=4)
rw_model.fit(X_train_clean, y_train)

# === PREDICT ===
rw_preds, rw_conf = rw_model.predict_with_confidence(X_test_clean)

# Ensure predictions are a list for evaluation
rw_preds = np.array(rw_preds)

# === EVALUATE ===
print("Random Wheel Accuracy: {:.5f}".format(accuracy_score(y_test, rw_preds)))
print("Confusion Matrix:")
print(pd.DataFrame(confusion_matrix(y_test, rw_preds)))

# === EXPLAIN A SAMPLE ===
sample_index = 0  # Change this to explain a different sample
sample_features = X_test_clean.iloc[sample_index]
explanation = rw_model.explain_instance(sample_features)
print("\nExplanation for sample index", sample_index)
for feature, importance in explanation.items():
    print(f"{feature}: {importance:.4f}")
print(f"\nPrediction: {rw_preds[sample_index]}, Confidence: {rw_conf[sample_index]:.4f}")
```

## Enhancing the Prediction

```python
from sklearn.base import BaseEstimator, ClassifierMixin
from sklearn.ensemble import StackingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.tree import DecisionTreeClassifier
import numpy as np
import pandas as pd

# ---- Random Wheel Classifier ----
class RandomWheelClassifier:
    def __init__(self, n_estimators=10, max_depth=5, random_state=42):
        self.n_estimators = n_estimators
        self.max_depth = max_depth
        self.random_state = random_state
        self.models = []

    def fit(self, X, y):
        self.models = []
        np.random.seed(self.random_state)
        for i in range(self.n_estimators):
            idx = np.random.choice(len(X), len(X), replace=True)
            X_sample = X.iloc[idx]
            y_sample = y[idx]
            clf = DecisionTreeClassifier(max_depth=self.max_depth, random_state=self.random_state + i)
            clf.fit(X_sample, y_sample)
            self.models.append(clf)

    def predict(self, X):
        preds = np.array([model.predict(X) for model in self.models])
        return np.round(np.mean(preds, axis=0)).astype(int)

    def predict_proba(self, X):
        preds = np.array([model.predict(X) for model in self.models])
        avg_probs = np.mean(preds, axis=0)
        return np.vstack([1 - avg_probs, avg_probs]).T

    def explain_instance(self, X_row):
        explanations = []
        for model in self.models:
            importances = model.feature_importances_
            explanations.append(importances)
        avg_importance = np.mean(explanations, axis=0)
        return dict(zip(X_row.index, avg_importance))

# ---- scikit-learn compatible wrapper ----
class RandomWheelWrapper(BaseEstimator, ClassifierMixin):
    def __init__(self, n_estimators=10, max_depth=5, random_state=42):
        self.n_estimators = n_estimators
        self.max_depth = max_depth
        self.random_state = random_state
        self.model = RandomWheelClassifier(n_estimators=n_estimators,
                                           max_depth=max_depth,
                                           random_state=random_state)

    def fit(self, X, y):
        self.model.fit(X, y)
        self.classes_ = np.unique(y)
        return self

    def predict(self, X):
        return self.model.predict(X)

    def predict_proba(self, X):
        return self.model.predict_proba(X)

    def get_params(self, deep=True):
        return {
            "n_estimators": self.n_estimators,
            "max_depth": self.max_depth,
            "random_state": self.random_state
        }

    def set_params(self, **params):
        for param, value in params.items():
            setattr(self, param, value)
        self.model = RandomWheelClassifier(**params)
        return self


# ---- Preprocess: Drop problematic columns if necessary ----
if 'wkphone' in X_train.columns:
    X_train_clean = X_train.drop(columns=['wkphone'])
    X_test_clean = X_test.drop(columns=['wkphone'])
else:
    X_train_clean = X_train.copy()
    X_test_clean = X_test.copy()

# ---- Initialize base models ----
rw_wrapper = RandomWheelWrapper(n_estimators=15, max_depth=4, random_state=42)

lgb_model = LGBMClassifier(num_leaves=31, max_depth=8, learning_rate=0.02, n_estimators=250,
                           subsample=0.8, colsample_bytree=0.8)

xgb_model = XGBClassifier(max_depth=12, n_estimators=250, min_child_weight=8,
                          subsample=0.8, learning_rate=0.02, seed=42)

# ---- Build stacking ensemble ----
estimators = [
    ('rw', rw_wrapper),
    ('lgb', lgb_model),
    ('xgb', xgb_model)
]

stack_model = StackingClassifier(
    estimators=estimators,
    final_estimator=LogisticRegression(),
    stack_method='predict_proba',
    passthrough=True,
    cv=5
)

# ---- Train and evaluate ----
stack_model.fit(X_train_clean, y_train)
y_pred_stack = stack_model.predict(X_test_clean)

print(" Stacked Model Accuracy: {:.5f}".format(accuracy_score(y_test, y_pred_stack)))
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_stack))
```

This model: Maintains high predictive power (XGBoost) and adds interpretability (Random Wheel).
